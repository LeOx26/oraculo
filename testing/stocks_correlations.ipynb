```python
import yfinance as yf
import pandas as pd
import plotly as plt
import plotly.express as px
```

```python
def stock(symbol):
    return yf.Ticker(symbol).history(start='2021-09-01', end=('2022-09-01'))

appl_data = stock('aapl')
appl_data.head()
```

```python
appl_data['Close'].values
```

```python
appl_data['Close'].tail()
```

```python
amd_data = stock('amd')
appl_data['CloseAPPL'] = appl_data.Close
amd_data['CloseAMD'] = amd_data.Close
appl_amd_close_corr = pd.concat([appl_data.CloseAPPL, amd_data.CloseAMD], axis=1)
appl_amd_close_corr.describe()
```
