```python
from binance.spot import Spot

client = Spot()
client.time()
```

```python
import pandas as pd
from datetime import datetime
def klines(s, i, startT=None, endT=None):
    if startT == None:
        st = None
    else:
        st = int(datetime.timestamp(datetime.strptime(startT+" 00:00:00", "%d/%m/%Y %H:%M:%S")))*1000
    if endT == None:
        ed = None
    else:
        ed = int(datetime.timestamp(datetime.strptime(endT+" 00:00:00", "%d/%m/%Y %H:%M:%S")))*1000
    data = pd.DataFrame(client.klines(symbol=s, interval=i,limit=1000, startTime=st, endTime=ed))
    data[0] = pd.to_datetime(data[0], unit='ms')
    data[6] = pd.to_datetime(data[6], unit='ms')
    data.drop(11, inplace=True, axis=1)
    data.columns = ['start', 'open', 'high', 'low', 'close', 'vol', 'end', 'quoteAssetVol', 'trades', 'takerBuyBaseAssetVol', 'takerBuyQuoteAssetVol']
    return data
```

```python
link = pd.DataFrame(klines('LINKUSDT', '1m', None, '11/12/2022'))
link.head()
```

```python
eth = klines(
    'ETHUSDT',
    '1m',
    None,
    '4/12/2022'
)
eth.head()
```
